```python
import pickle
import pandas as pd
import numpy as np
import os
```

```python
dir_data = './data/baseline/'
```

```python
# loop over all files in the directory
all_auto_captions = 0
autocaptioned_data = pd.DataFrame()
for file in os.listdir(dir_data):
    if not file.endswith('.pickle'):
        continue
    # open the pickle file
    with open(dir_data + file, 'rb') as f:
        data = pickle.load(f)
    # count how many Video Transcript are not None
    auto_caption = data["Video Transcript"].apply(lambda x: x is not None)

    # subset the data to only the rows with auto captions
    data = data[auto_caption]
    autocaptioned_data = pd.concat([autocaptioned_data, data])
    # append the auto captions to the list
    all_auto_captions += auto_caption.sum()

print(f"Total auto captions: {all_auto_captions}")
```

```python
# extract year from video timestamp
autocaptioned_data['Video Timestamp'] = pd.to_datetime(autocaptioned_data['Video Timestamp'])
autocaptioned_data['Year'] = autocaptioned_data['Video Timestamp'].apply(lambda x: x.year)
autocaptioned_data
```

Clean text

```python
import re

def clean_text(text):

    # if text type is float or NoneType, return None
    if type(text) == float or type(text) == type(None):
        output = ""
        return output
    #replace . with space
    output = text.replace('.', ' ')

    #remove tags such as [Music]
    output = re.sub(r'\[.*?\]', '', output)

    output = output.lower()

    #remove string if string only contains punctuation
    if sum([i.isalpha() for i in output])== 0:
        output = None
        return output
        
    #remove if number of unique words is less than 5
    if len(set(output.split())) < 5:
        output = None
    
    return output
```

```python
# apply the function to the transcript column if it is not None
autocaptioned_data['Video Transcript Clean'] = autocaptioned_data['Video Transcript'].apply(clean_text)
# check how many transcripts are left
autocaptioned_data['Video Transcript Clean'].notnull().sum()

autocaptioned_data = autocaptioned_data[autocaptioned_data['Video Transcript Clean'].notnull()]
autocaptioned_data
```

```python
# get only videos after 2021
autocaptioned_data = autocaptioned_data[autocaptioned_data['Year'] >= 2021]
autocaptioned_data
```

```python
# save the data
autocaptioned_data.to_csv('./data/baseline/autocaptioned_data.csv', index=False)
```
