# TT

```python
for creator in dictionary_channelid.keys():

    df_all_tiktok = []
    for year in range(2021, 2024):
        with open("./data/search/creators/all_keywords_{}_{}-01-01.json".format(creator, year), "r") as f:
            df = pd.read_json(f)
            df_all_tiktok.append(df)
    
    df_tt = pd.concat(df_all_tiktok)

    # Clean description
    
    # tokenize Â¨video_descriptionÂ¨
    df_tt["video_description"] = df_tt["video_description"].apply(lambda x: x.split())


    # remove tokens containing Â¨httpÂ¨
    df_tt["video_description"] = df_tt["video_description"].apply(lambda x: [x[i] for i in range(len(x)) if ("http" not in x[i])])

    # if all tokens are hashtags or mentions, remove all tokens
    df_tt["video_description"] = df_tt["video_description"].apply(lambda x: [] if all([x[i][0] == "#" or x[i][0] == "@" for i in range(len(x))]) else x)

    # remove tokens containing Â¨#Â¨ from the end until the previous token does not contain Â¨#Â¨, do not remove if the token is the first one
    df_tt["video_description"] = df_tt["video_description"].apply(lambda x: [x[i] for i in range(len(x)) if (i==0 or ("#" not in x[i]) or ("#" not in x[i-1]))])
    #df_tt["video_description"] = df_tt["video_description"].apply(lambda x: [x[i] for i in range(len(x)) if (i==0 or ("#" not in x[i]) or ("#" not in x[i-1] and i!=len(x)-1))])
    # normalize tokens containing Â¨#Â¨ or Â¨@Â¨
    df_tt["video_description"] = df_tt["video_description"].apply(lambda x: [x[i].replace("#", "").replace("@", "") for i in range(len(x))])

    
    # remove the last token if the one before ends with a punctuation mark
    # if there are more than 2 tokens, remove the last token if the one before ends with a punctuation mark
    df_tt["video_description"] = df_tt["video_description"].apply(lambda x: x[:-1] if (len(x)>=2 and x[-2][-1] in [".", "!", "?"]) else x)

    # for i in range(len(df_tt)):
    #     print(df_tt["video_description"].iloc[i])
    #     print("")

    # join tokens back into sentences
    df_tt["video_description"] = df_tt["video_description"].apply(lambda x: " ".join(x))

    df_tt.reset_index(inplace=True)


    # print all descriptions in a readable format
    for i in range(len(df_tt)):
        print(df_tt["video_description"].iloc[i])
        print("")

    #save as pickle
    df_tt.to_pickle("./data/search/creators/all_keywords_{}_2021-2023.pkl".format(creator))
```
