```python
df = pd.read_csv("/Users/arpe/Documents/ITU/COCOONS/tiktok/cos_sim_creators_withtrans_newdata.csv")
df
```

```python
# plot the cosine similarity, mean and std per creator
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(rc={'figure.figsize':(11.7,8.27)})

# compare avg_cos_sim_yt and avg_cos_sim_tt per creator with histogram
df = df.melt(id_vars=["creator"], value_vars=["avg_cos_sim_yt", "avg_cos_sim_tt"])
df = df.rename(columns={"variable": "platform", "value": "cosine similarity"})
df["platform"] = df["platform"].str.replace("avg_cos_sim_", "")
df["platform"] = df["platform"].str.upper()
df["platform"] = df["platform"].str.replace("TT", "TIKTOK")
df["platform"] = df["platform"].str.replace("YT", "YOUTUBE")
df["platform"] = df["platform"].str.replace("_", " ")
df["platform"] = df["platform"].str.title()

ax = sns.barplot(x="creator", y="cosine similarity", hue="platform", data=df)
plt.xticks(rotation=90)
plt.show()

```

```python
# open descriptions_validate
df_tt = pd.read_csv("/Users/arpe/Documents/ITU/COCOONS/tiktok/descriptions_validate_tt_newdata.csv")
df_yt = pd.read_csv("/Users/arpe/Documents/ITU/COCOONS/tiktok/descriptions_validate_yt_newdata.csv")

# merge the two dataframes, add tt and yt as platform
df = pd.merge(df_tt, df_yt, on="creator", how="outer", suffixes=("_tt", "_yt"))
df
```

```python
# get avg cossim tt and yt overall
print("Avg. cossim desc-transcript overall TT", df["avg_cos_sim_desc_transc_tt"].mean())
print("Avg. cossim desc-transcript overall YT", df["avg_cos_sim_desc_transc_yt"].mean())
```

```python
# get avg std cossim tt and yt overall
print("Std. cossim desc-transcript overall TT", df["std_cos_sim_desc_transc_tt"].mean())
print("Std. cossim desc-transcript overall YT", df["std_cos_sim_desc_transc_yt"].mean())
```

```python
# plot the cosine similarity, mean and std per creator
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(rc={'figure.figsize':(11.7,8.27)})

# compare avg_cos_sim_yt and avg_cos_sim_tt per creator with histogram
df = df.melt(id_vars=["creator"], value_vars=["avg_cos_sim_desc_transc_yt", "avg_cos_sim_desc_transc_tt"])
df = df.rename(columns={"variable": "platform", "value": "cosine similarity"})
df["platform"] = df["platform"].str.replace("avg_cos_sim_desc_transc_", "")
df["platform"] = df["platform"].str.upper()
df["platform"] = df["platform"].str.replace("TT", "TIKTOK")
df["platform"] = df["platform"].str.replace("YT", "YOUTUBE")
df["platform"] = df["platform"].str.replace("_", " ")
df["platform"] = df["platform"].str.title()

ax = sns.barplot(x="creator", y="cosine similarity", hue="platform", data=df)
plt.xticks(rotation=90)
plt.title("Cosine similarity between description and corresponding video transcription, by user")
plt.show()
```
